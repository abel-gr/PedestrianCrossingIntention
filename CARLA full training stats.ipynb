# Comparison between the amounts of frames of the temporal model of the Graph Convolutional Neural Network

For Pedestrian Crossing Intention binary classification.

## Read the data

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

import os
```

Read the already saved metrics:

```python
path = 'exportedModels/CARLA/Full train dataset/'

dataframes = {}
for model in os.listdir(path):
    train_csv = pd.read_csv(path + model + '/train_metrics.csv', index_col=0)
    val_csv = pd.read_csv(path + model + '/val_metrics.csv', index_col=0)
    test_csv = pd.read_csv(path + model + '/test_metrics.csv', index_col=0)
    
    dataframes[int(model.split(" ")[2])] = {
        "train": train_csv,
        "val": val_csv,
        "test": test_csv
    }
```

Example of train metrics for the Spatial-Temporal model that uses 2 frames:

```python
dataframes[2]['train']
```

## Plots and metrics

```python
modelsMetricDic = {}

for datasubset in ["train", "val", "test"]:
    
    print(datasubset + " metrics:\n\n")

    metrics = list(dataframes[2][datasubset].keys())

    for metric in metrics:
        if metric.lower() == 'epoch':
            continue
            
        print('Metric: ' + metric)

        metricValues = []
        numbersofFrames = []

        # Iterate over the different saved models metrics
        for numFrames, dfs in dataframes.items():
            df = dfs[datasubset]

            if 'time' in metric:
                metricValue = np.mean(df[metric].to_numpy())
            else:  
                metricValue = df.iloc[-1][metric]

            numbersofFrames.append(numFrames)
            metricValues.append(round(metricValue, 4))
            
            if 'time' not in metric.lower() and 'loss' not in metric.lower():
                if numFrames in modelsMetricDic:
                    if datasubset in modelsMetricDic[numFrames]:
                        modelsMetricDic[numFrames][datasubset][metric] = round(metricValue, 4)
                    else:
                        modelsMetricDic[numFrames][datasubset] = {metric: round(metricValue, 4)}                        
                else:
                    modelsMetricDic[numFrames] = {datasubset: {metric: round(metricValue, 4)}}

        title = metric + " regarding the amount of video frames \nused in the Spatial-Temporal model ("
        title = title + datasubset + " subset)"
        
        minY = np.min(metricValues)*0.99
        maxY = np.max(metricValues)*1.01

        fig, ax = plt.subplots(figsize=(10,6))
        ax.set_title(title, size=16)
        ax.set_xlabel("Number of frames", size=14)
        ax.set_ylabel(metric, size=14)
        ax.set_xticks(numbersofFrames)
        ax.set_yticks(np.round(np.arange(minY, maxY, (maxY-minY)/10), 4))
        ax.set_ylim(minY, maxY)
        ax.bar(numbersofFrames, metricValues)
        plt.show()
                
        metric_df = pd.DataFrame({'Number of frames': numbersofFrames, metric: metricValues})
        
        print(metric_df)
        print('\n---------------------------------------\n\n\n\n')
        
    print('\n\n\n\n-------------------------------------------------------------\n\n')
```
